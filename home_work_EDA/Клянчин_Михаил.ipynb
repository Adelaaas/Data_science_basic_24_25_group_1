## Загрузка данных

Вам дан датасет о качестве воздуха в различных регионах: `pollution_dataset.csv`

Основные характеристики:

- Температура (°C): Средняя температура региона.
- Влажность (%): Относительная влажность, зарегистрированная в регионе.
- Концентрация PM2.5 (мкг/м³): Уровни мелких твердых частиц.
- Концентрация PM10 (мкг/м³): Уровни крупных твердых частиц.
- Концентрация NO2 (ppb): Уровни диоксида азота.
- Концентрация SO2 (ppb): Уровни диоксида серы.
- Концентрация CO (ppm): Уровни оксида углерода.
- Близость к промышленным зонам (км): Расстояние до ближайшей промышленной зоны.
- Плотность населения (чел./км²): Количество людей на квадратный километр в регионе.
- Уровни качества воздуха:
    - Хорошее (Good): Чистый воздух с низким уровнем загрязнения.
    - Умеренный (Moderate): Приемлемое качество воздуха, но присутствуют некоторые загрязняющие вещества.
    - Плохой (Poor): Заметное загрязнение, которое может вызвать проблемы со здоровьем у чувствительных групп.
    - Опасный (Hazardous): Сильно загрязненный воздух, представляющий серьезные риски для здоровья населения.

```python
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv(r"https://raw.githubusercontent.com/Adelaaas/Data_science_basic_24_25_group_1/refs/heads/main/home_work_EDA/pollution_dataset.csv")
df
```

## График 1

Построить точечный график зависимости переменной `NO2` от `CO` с учётом третей переменной `Air Quality`.

```python
x = df["NO2"]
y = df["CO"]
z = df['Air Quality']
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x, y=y, hue=z s=10)
plt.title("NO2 by CO with Air Quality")
plt.xlabel("NO2")
plt.ylabel("CO")
plt.grid(True)
plt.legend(title="Air Quality")
plt.show()
```

Можно количественно понять соотношение CO и NO2.
К тому же видны определенные кластеры. Однако можно заметить выбросы вроде точки Poor в зеленом кластере (Good)

## График 2

Постройте 4 графика на одной фигуре, как расположить координатные плоскости - это на ваше усмотрение.

Это должны быть графики для следующих переменных:
- `Air Quality` - гистограмма
- `Population_Density` - график распределения
- `Temperature` - график распределения
- `Humidity` - график распределения

Каждый график должен быть своего цвета, также задайте цвет отличный от белого для бекграунда всей фигуры.

```python
plt.figure(figsize=(12, 6))

ax1 = plt.subplot(2, 2, 1)
ax1.set_facecolor('lightgray')
plt.hist(df["Air Quality"], bins=5, color='blue', edgecolor='black', rwidth=0.9)
plt.title("Air Quality")
plt.xlabel("Rate")
plt.ylabel("Number")
plt.tight_layout()

value_counts = df['Population_Density'].value_counts().reset_index()
value_counts.columns = ['Population_Density', 'Count']
ax2 = plt.subplot(2, 2, 2)
ax2.set_facecolor('lightgray')
plt.title("Population Density")
sns.barplot(data=value_counts, x='Count', y='Population_Density', palette=["red", "yellow"])
plt.tight_layout()

ax3 = plt.subplot(2, 2, 3)
ax3.set_facecolor('lightgray')
plt.title("Temperature")
sns.barplot(data=df, y='Temperature', x=df["Temperature"].value_counts(), palette=["green", "blue"])
plt.xticks([1, 5, 10, 20, 30], [1, 5, 10, 20, 30])
plt.xlabel("Count")
plt.tight_layout()

ax4 = plt.subplot(2, 2, 4)
ax4.set_facecolor('lightgray')
plt.title("Humidity")
sns.barplot(data=df, y='Humidity', x=df["Humidity"].value_counts(), palette=["pink", "gray"])
plt.xticks([1, 5, 10, 20], [1, 5, 10, 20])
plt.xlabel("Count")
plt.tight_layout()

plt.subplots_adjust(hspace=0.5, wspace=0.4)
plt.show()
```

На первом графике видно что больше всего хорошее качество воздуха, встречается 2000 раз.
На втором можно заметить что максимальная плотность - 800, минимальная - около 400.
На третьем видно  максимум и минимум: 18 градусов и 50 соотвественно.
На четвертом графике можно заметить максимум и минимум влажности: 50 и 100 процентов соответственно.

## График 3

Визуализируйте матрицу корреляции в виде heatmap.

Все должно быть ЧИТАБЕЛЬНО!

```python
data = {
    'Air Quality': ['Good', 'Moderate', 'Poor', 'Hazardous', 'Good', 'Poor']
}
df2 = df
df2['Air Quality'] = df['Air Quality'].replace({
    'Good': 1,
    'Moderate': 2,
    'Poor': 3,
    'Hazardous': 4
})
matrix = df2.corr()
plt.figure(figsize=(8, 6))
plt.title("Corr Matrix")
sns.heatmap(matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```

На матрице можно заметить большой процент корреляции между PM10  и PM2.5 и маленькую зависимость между  CO и Proximity_to_Industrial_Areas

## График 5

Постройте heatmap для отображения пропусков в данных.

```python
df.replace("", float("NaN"), inplace=True)
df_with_missing = df[df.isnull().any(axis=1)]
plt.figure(figsize=(12, 10))
sns.heatmap(df_with_missing.isnull(), cbar=False, cmap='viridis', linewidths=0.5, linecolor='white')
plt.ylabel("Line number")
plt.xlabel("Column")
plt.title("Voids heatmap")
plt.show()
```

Можно заметить количество пропусков на каждой строке:
* PM2.5 - 20  
*   SO2 - 5
*   PM10 - 1






## График 6

Постройте график среднего значения переменной `PM10` в зависимости от переменной `Air Polution`.

```python
a =  df.groupby('Air Quality')['PM10'].mean().reset_index()
plt.figure(figsize=(10, 6))
category_order = ['Good', 'Moderate', 'Poor', 'Hazardous']
sns.barplot(data=a, x='Air Quality', y='PM10', palette='Reds',order = category_order)
```

Можно заметить что с возрастанием среднего количества PM10 ухудшается качество воздуха
