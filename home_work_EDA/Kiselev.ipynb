```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('weatherHistory_var1.csv')

#print(data.head())

#print(data[['Precip Type', 'Season', 'DayOfWeek', 'Year']].isnull().sum())

data = data.dropna(subset=['Precip Type', 'Season', 'DayOfWeek', 'Year'])

sns.set(style="darkgrid")

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.patch.set_facecolor('#f0f0f0')

# Precip Type
sns.countplot(ax=axes[0, 0], x='Precip Type', data=data, palette='viridis')
axes[0, 0].set_title('   ', fontsize=14)
axes[0, 0].set_xlabel(' ', fontsize=12)
axes[0, 0].set_ylabel('', fontsize=12)

# Season
sns.countplot(ax=axes[0, 1], x='Season', data=data, palette='plasma')
axes[0, 1].set_title('  ', fontsize=14)
axes[0, 1].set_xlabel('', fontsize=12)
axes[0, 1].set_ylabel('', fontsize=12)

# DayOfWeek
sns.countplot(ax=axes[1, 0], x='DayOfWeek', data=data, palette='magma')
axes[1, 0].set_title('   ', fontsize=14)
axes[1, 0].set_xlabel(' ', fontsize=12)
axes[1, 0].set_ylabel('', fontsize=12)

# Year
sns.countplot(ax=axes[1, 1], x='Year', data=data, palette='coolwarm')
axes[1, 1].set_title('  ', fontsize=14)
axes[1, 1].set_xlabel('', fontsize=12)
axes[1, 1].set_ylabel('', fontsize=12)

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('weatherHistory_var1.csv')

#print(data.head())

numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns

#print(data[numeric_columns].isnull().sum())

data = data.dropna(subset=numeric_columns)

corr_matrix = data[numeric_columns].corr()



#print(corr_matrix)

sns.set(style="white")

plt.figure(figsize=(12, 8))

sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    vmin=-1, vmax=1,
    linewidths=0.5,
    cbar=True
)

plt.title('   ', fontsize=16)
plt.xlabel('', fontsize=14)
plt.ylabel('', fontsize=14)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('weatherHistory_var1.csv')

#print(data.head())

#print(data[['Temperature (C)', 'Pressure (millibars)', 'Precip Type']].isnull().sum())

data = data.dropna(subset=['Temperature (C)', 'Pressure (millibars)', 'Precip Type'])

#print(data['Precip Type'].unique())

sns.set(style="whitegrid")

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.boxplot(ax=axes[0], x='Precip Type', y='Temperature (C)', data=data, palette='coolwarm')
axes[0].set_title('    ', fontsize=14)
axes[0].set_xlabel(' ', fontsize=12)
axes[0].set_ylabel(' (C)', fontsize=12)

sns.boxplot(ax=axes[1], x='Precip Type', y='Pressure (millibars)', data=data, palette='magma')
axes[1].set_title('    ', fontsize=14)
axes[1].set_xlabel(' ', fontsize=12)
axes[1].set_ylabel(' (millibars)', fontsize=12)

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('weatherHistory_var1.csv')

data['Formatted Date'] = pd.to_datetime(data['Formatted Date'])
data['Year'] = data['Formatted Date'].dt.year


average_temp_per_year = data.groupby('Year')['Temperature (C)'].mean().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(average_temp_per_year['Year'], average_temp_per_year['Temperature (C)'], marker='o', linestyle='-', color='g')
plt.title('Average Annual Temperature Over Years')
plt.xlabel('Year')
plt.ylabel('Average Temperature (C)')
plt.grid(True)
plt.show()
```
