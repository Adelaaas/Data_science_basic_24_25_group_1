```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('weatherHistory_var1.csv')

#print(data.head())

#print(data[['Temperature (C)', 'Wind Speed (km/h)', 'Precip Type']].isnull().sum())

data = data.dropna(subset=['Temperature (C)', 'Wind Speed (km/h)', 'Precip Type'])
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Wind Speed (km/h)', y='Temperature (C)', hue='Precip Type', data=data, palette='magma', alpha=0.5)

plt.title('Зависимость температуры от скорости ветра с учетом типа осадков', fontsize=16)
plt.xlabel('Скорость ветра (km/h)', fontsize=14)
plt.ylabel('Температура (C)', fontsize=14)

plt.legend(title='Тип осадков')

plt.show()
```

По данному графику можно понять, что корреляция между скоростью ветра и температурой почти отсутсвует, мы можем сделать выводы о том, что при увеличении скорости ветра температура уменьшается по модулю.График показывает, что тип осадков (дождь или снег) также влияет на распределение точек. Например, при низких температурах чаще встречаются осадки в виде снега, а при более высоких — в виде дождя.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('weatherHistory_var1.csv')

#print(data.head())

#print(data[['Precip Type', 'Season', 'DayOfWeek', 'Year']].isnull().sum())

data = data.dropna(subset=['Precip Type', 'Season', 'DayOfWeek', 'Year'])

sns.set(style="darkgrid")

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.patch.set_facecolor('#f0f0f0')

# Precip Type
sns.countplot(ax=axes[0, 0], x='Precip Type', data=data, palette='viridis')
axes[0, 0].set_title('Распределение по типу осадков', fontsize=14)
axes[0, 0].set_xlabel('Тип осадков', fontsize=12)
axes[0, 0].set_ylabel('Количество', fontsize=12)

# Season
sns.countplot(ax=axes[0, 1], x='Season', data=data, palette='plasma')
axes[0, 1].set_title('Распределение по сезонам', fontsize=14)
axes[0, 1].set_xlabel('Сезон', fontsize=12)
axes[0, 1].set_ylabel('Количество', fontsize=12)

# DayOfWeek
sns.countplot(ax=axes[1, 0], x='DayOfWeek', data=data, palette='magma')
axes[1, 0].set_title('Распределение по дням недели', fontsize=14)
axes[1, 0].set_xlabel('День недели', fontsize=12)
axes[1, 0].set_ylabel('Количество', fontsize=12)

# Year
sns.countplot(ax=axes[1, 1], x='Year', data=data, palette='coolwarm')
axes[1, 1].set_title('Распределение по годам', fontsize=14)
axes[1, 1].set_xlabel('Год', fontsize=12)
axes[1, 1].set_ylabel('Количество', fontsize=12)

plt.tight_layout()

plt.show()
```

Гистограмма для Precip Type:
Осадков в виде дождя выпадает гораздо больше, чем в виде снега.

Гистограмма для Season:
Летом было собрано больше всего данных, вероятно это связанос с более активной погодой в это время

Гистограмма для DayOfWeek:
В выходные дни данных меньше, это может быть связано с особенностями сбора данных.

Гистограмма для Year:
График показывает распределение данных по годам.
В последние годы данных больше, это может указывать на улучшение системы сбора данных.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('weatherHistory_var1.csv')

#print(data.head())

numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns

#print(data[numeric_columns].isnull().sum())

data = data.dropna(subset=numeric_columns)

corr_matrix = data[numeric_columns].corr()



#print(corr_matrix)

sns.set(style="white")

plt.figure(figsize=(12, 8))

sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    vmin=-1, vmax=1,
    linewidths=0.5,
    cbar=True
)

plt.title('Матрица корреляции числовых переменных', fontsize=16)
plt.xlabel('Переменные', fontsize=14)
plt.ylabel('Переменные', fontsize=14)
plt.show()
```

Корреляционная матрица показывает корреляцию между некоторыми праметрами. 1 - полная корреляция. -1 - отрицательная корреляция. 0 - корреляция отсутсвует.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('weatherHistory_var1.csv')

#print(data.head())

#print(data[['Temperature (C)', 'Pressure (millibars)', 'Precip Type']].isnull().sum())

data = data.dropna(subset=['Temperature (C)', 'Pressure (millibars)', 'Precip Type'])

#print(data['Precip Type'].unique())

sns.set(style="whitegrid")

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.boxplot(ax=axes[0], x='Precip Type', y='Temperature (C)', data=data, palette='coolwarm')
axes[0].set_title('Распределение температуры по типу осадков', fontsize=14)
axes[0].set_xlabel('Тип осадков', fontsize=12)
axes[0].set_ylabel('Температура (C)', fontsize=12)

sns.boxplot(ax=axes[1], x='Precip Type', y='Pressure (millibars)', data=data, palette='magma')
axes[1].set_title('Распределение давления по типу осадков', fontsize=14)
axes[1].set_xlabel('Тип осадков', fontsize=12)
axes[1].set_ylabel('Давление (millibars)', fontsize=12)

plt.tight_layout()

plt.show()
```

График распределения температуры по типу осадеов показывает: 1 - максимальныно допустимую температуру для данного вида осадков. 2 - среднюю температуру для данного вида осадков.

```python

data = pd.read_csv('weatherHistory_var1.csv')

data['Formatted Date'] = pd.to_datetime(data['Formatted Date'])

plt.figure(figsize=(12, 6))
sns.histplot(data=data, x='Wind Bearing (degrees)', hue='Precip Type', kde=True, palette='coolwarm')
plt.title('Распределение силы ветра по типу осадков')
plt.xlabel('Направление ветра в градусах')
plt.ylabel('Частота')
plt.grid(True)
plt.show()
```

График показывает, как направление ветра (Wind Bearing (degrees)) распределено для разных типов осадков (Precip Type).

Например, если для дождя (rain) направление ветра сосредоточено в определенном диапазоне (например, 0–180 градусов), это может указывать на преобладание ветров с определенного направления во время дождя.

Различия между типами осадков:

Если распределения для дождя и снега значительно отличаются, это может указывать на разные погодные условия, связанные с этими типами осадков.
Снег может чаще сопровождаться ветрами с северного направления, а дождь — с южного.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('weatherHistory_var1.csv')

data['Formatted Date'] = pd.to_datetime(data['Formatted Date'])
data['Year'] = data['Formatted Date'].dt.year


average_temp_per_year = data.groupby('Year')['Temperature (C)'].mean().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(average_temp_per_year['Year'], average_temp_per_year['Temperature (C)'], marker='o', linestyle='-', color='g')
plt.title('Average Annual Temperature Over Years')
plt.xlabel('Year')
plt.ylabel('Average Temperature (C)')
plt.grid(True)
plt.show()
```

График показывает рост средней температуры с течением времени, это может указывать на глобальное потепление.
Если температура колеблется без явного тренда, это может говорить о стабильности климатических условий.



```python

import pandas as pd
import matplotlib.pyplot as plt

# Группируем данные по часам и находим среднюю видимость
hourly_visibility = data.groupby('Hour')['Visibility (km)'].mean()

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(hourly_visibility.index, hourly_visibility.values, marker='o', linestyle='-', color='g')
plt.title('Средняя видимость по часам суток')
plt.xlabel('Час суток')
plt.ylabel('Средняя видимость (км)')
plt.grid(True)
plt.show()
```
