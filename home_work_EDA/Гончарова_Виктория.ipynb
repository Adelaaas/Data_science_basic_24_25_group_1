# Задача 1

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
df = pd.read_csv('sales_dataset.csv')
plt.figure(figsize=(20, 10))
sns.scatterplot(df, x='price', y='Area_in_Marla', hue='purpose')
plt.title('Зависимость площади продаваемого и сдающегося в аренду жилья от цены на него')
plt.show()
```

Вывод: цены на съёмное жильё возрастают в зависимотси от площади, но меняются в небольшом диапазоне и малы по сравнению с ценами на продаваемое. Цены на продаваемое жильё слабо зависят от площади жилья, в основном продаётся жильё площадью до 1000 марла по ценой до 250 000 000 рупий(?).  

# Задача 2

```python
fig, axes = plt.subplots(2, 2, figsize=(20, 20))
sns.countplot(x = df['property_type'], ax=axes[0, 0], hue=df['property_type'], palette='Set2', legend=False)
axes[0, 0].set_title('Количественное соотношение жилья разных типов')
sns.countplot(x = df['city'], ax=axes[0, 1], hue=df['city'], legend=False, palette='dark:#bdbdf2')
axes[0, 1].set_title('Количественнное соотношение жилья в разных городах')
sns.countplot(x = df['purpose'], ax=axes[1, 0], hue=df['purpose'], palette='hls', legend=False)
axes[1, 0].set_title('Количественное соотношение продаваемого и сдающегося жилья')
sns.countplot(x = df['bedrooms'], ax=axes[1, 1], hue=df['bedrooms'], palette='Set3', legend=False)
axes[1, 1].set_title('Количественное соотношение жилья с разным количеством спален')
fig.set_facecolor('#f5f5f5')
plt.show()
```

Вывод: 1) чаще всего предлагают частные дома, реже -- квартиры, реже всего - пентхаусы; 2) больше всего предлагают жильё в Карачи, меньше всего -- в Фаисалабаде; 3) почти в три раза больше жилья продаётся, чем сдаётся в аренду; 4) чаще всего предлагаются дома с тремя спальнями, вариантов жилья с 10 и более спальнями меньше 100.

# Задача 5




Постройте heatmap для отображения пропусков в данных.

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

cmap = sns.cubehelix_palette(as_cmap=True, light=.9)

sns.heatmap(df.isna().transpose(), cmap=cmap,
            cbar_kws={'label': 'Missing Data'}, ax=axes[0])
axes[0].set_title('Heatmap пропусков в данных')

sns.histplot(data=df.isna().melt(value_name="missing"), y="variable", hue="missing",
             multiple="fill", ax=axes[1], palette="viridis")
axes[1].set_title('Доля пропусков по столбцам', fontsize=16)
axes[1].set_xlabel('Доля пропусков', fontsize=14)
axes[1].set_ylabel('Столбцы', fontsize=14)

plt.tight_layout()
plt.show()
```

**Вывод:**
По двум представленным графикаи мы видим, что пропусков данных нет ни в одной категории


# Задача 6

Постройте график среднего значения переменной `price` в зависимости от переменной `Season`.

```python



fig, axes = plt.subplots(1, 2, figsize=(20, 8))
sns.barplot(data=df.groupby('Season')['price'].mean().reset_index(), x='Season', y='price', palette='viridis', ax=axes[0])


axes[0].set_title('Среднее значение price в зависимости от Season', fontsize=16)
axes[0].set_xlabel('Season', fontsize=14)
axes[0].set_ylabel('Среднее значение price', fontsize=14)

sns.barplot(data=df, x='Season', y='price', palette='viridis', ci='sd', ax = axes[1])
axes[1].set_title('Стандартное отклонение от среднего значения price в зависимости от Season', fontsize=14)
axes[1].set_xlabel('Season', fontsize=14)
axes[1].set_ylabel('Среднее значение price', fontsize=14)

plt.show()
```

**Выводы:**

1.    По первому графику мы видим, что средняя цена в каждом из сезонов различается. В сезон 2 среднее значение цены максимально, а в сезон 4 минимально. Следовательно наиболее выгодно продавать и сдавать недвижиость в сезон 2, а покупать и заключать договоры аренды в сезон 4.
2.   Второй график показывает стандартное отлонение от среднего значения. Размер черных усов отображает разброс цен. Мы видим, что в сезон 4 цены наиболее приближены к среднему знаечнию, а в сезон 2 разброс максимальный.

Первый график -- 3D зависимость трех данных (цена, тип, кол-во спален). Второй график -- Joy Plot. Отличный способ визуализировать распределение большего числа групп по отношению друг к другу. По графику наглядно видно распределение количества спален в разных типах недвижимости. Такой график упрощает сравнение типов по количеству спален.
