```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных (предполагается, что файл 'pollution_dataset.csv' доступен)
data = pd.read_csv('pollution_dataset.csv')

# Выбор числовых столбцов
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns

# Проверка на пропуски и их удаление
print(data[numeric_columns].isnull().sum())
data = data.dropna(subset=numeric_columns)

# Построение матрицы корреляции
correlation = data[numeric_columns].corr()

# Настройка размера фигуры
plt.figure(figsize=(12, 8))

# Используем стиль seaborn для белой сетки
sns.set(style="whitegrid")

# Построение heatmap с измененной цветовой палитрой
sns.heatmap(correlation, annot=True, fmt=".2f", cmap='RdBu_r',
            cbar=True, square=True, linewidths=0.5,
            linecolor='white', vmin=-1, vmax=1)

# Настройка заголовка
plt.title('Матрица корреляции качества воздуха', fontsize=16)

# Поворот подписей осей с черным цветом
plt.xticks(rotation=45, fontsize=12, fontweight='normal', color='black')
plt.yticks(rotation=45, fontsize=12, fontweight='normal', color='black')

# Показ графика
plt.tight_layout()  # Подгонка всех элементов графика
plt.show()


```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
# Загрузка данных (предполагается, что файл 'pollution_dataset.csv' доступен)

data = pd.read_csv('pollution_dataset.csv')


# Убедимся, что 'Air Quality' — это категориальный тип
data['Air Quality'] = data['Air Quality'].astype('category')

# Построение графика
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Air Quality', y='PM10', palette='Blues')
plt.title('Среднее значение PM10 в зависимости от качества воздуха', fontsize=16)
plt.xlabel('Качество воздуха', fontsize=12)
plt.ylabel('Концентрация PM10 (мкг/м³)', fontsize=12)
plt.tight_layout()  # Подгонка отображения для более чистого вида
plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import find_peaks

# Добавляем временной ряд
start_date = "2020-01-01"
data["Day"] = pd.date_range(start=start_date, periods=len(data), freq="D")

# Находим пики и впадины
peaks, _ = find_peaks(data["PM10"])  # Пики
valleys, _ = find_peaks(-data["PM10"])  # Впадины

# Построение графика
plt.figure(figsize=(12, 6))

# Строим временной ряд
plt.plot(data["Day"], data["PM10"], label="Концентрация PM10", color="b", marker="o", markersize=2, linewidth=1)

# Отмечаем пики
plt.plot(data["Day"].iloc[peaks], data["PM10"].iloc[peaks], "rx", label="Пики")

# Отмечаем впадины
plt.plot(data["Day"].iloc[valleys], data["PM10"].iloc[valleys], "go", label="Впадины")

# Добавляем подписи
plt.title("Временные ряды PM10 с пиками и впадинами", fontsize=16)
plt.xlabel("Дата", fontsize=12)
plt.ylabel("Концентрация PM10 (мкг/м³)", fontsize=12)
plt.xticks(rotation=45)
plt.legend()

# Показать график
plt.tight_layout()
plt.show()



```
