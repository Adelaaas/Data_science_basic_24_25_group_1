```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#   (,   'pollution_dataset.csv' )
data = pd.read_csv('pollution_dataset.csv')

#   
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns

#      
print(data[numeric_columns].isnull().sum())
data = data.dropna(subset=numeric_columns)

#   
correlation = data[numeric_columns].corr()

#   
plt.figure(figsize=(12, 8))

#   seaborn   
sns.set(style="whitegrid")

#  heatmap    
sns.heatmap(correlation, annot=True, fmt=".2f", cmap='RdBu_r',
            cbar=True, square=True, linewidths=0.5,
            linecolor='white', vmin=-1, vmax=1)

#  
plt.title('   ', fontsize=16)

#      
plt.xticks(rotation=45, fontsize=12, fontweight='normal', color='black')
plt.yticks(rotation=45, fontsize=12, fontweight='normal', color='black')

#  
plt.tight_layout()  #    
plt.show()


```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
#   (,   'pollution_dataset.csv' )

data = pd.read_csv('pollution_dataset.csv')


# ,  'Air Quality'    
data['Air Quality'] = data['Air Quality'].astype('category')

#  
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Air Quality', y='PM10', palette='Blues')
plt.title('  PM10     ', fontsize=16)
plt.xlabel(' ', fontsize=12)
plt.ylabel(' PM10 (/)', fontsize=12)
plt.tight_layout()  #      
plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import find_peaks

#   
start_date = "2020-01-01"
data["Day"] = pd.date_range(start=start_date, periods=len(data), freq="D")

#    
peaks, _ = find_peaks(data["PM10"])  # 
valleys, _ = find_peaks(-data["PM10"])  # 

#  
plt.figure(figsize=(12, 6))

#   
plt.plot(data["Day"], data["PM10"], label=" PM10", color="b", marker="o", markersize=2, linewidth=1)

#  
plt.plot(data["Day"].iloc[peaks], data["PM10"].iloc[peaks], "rx", label="")

#  
plt.plot(data["Day"].iloc[valleys], data["PM10"].iloc[valleys], "go", label="")

#  
plt.title("  PM10    ", fontsize=16)
plt.xlabel("", fontsize=12)
plt.ylabel(" PM10 (/)", fontsize=12)
plt.xticks(rotation=45)
plt.legend()

#  
plt.tight_layout()
plt.show()



```
