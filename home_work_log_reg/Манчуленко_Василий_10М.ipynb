## График 1

Построить точечный график зависимости переменной `Area_in_Marla` от `price` с учётом третей переменной `purpose`.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('sales_dataset.csv')
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Area_in_Marla', y='price', hue='purpose', alpha=0.7)
plt.title('Зависимость Area_in_Marla от price с учетом purpose')
plt.xlabel('Area in Marla')
plt.ylabel('Price')
plt.legend(title='Purpose')
plt.grid(True)
plt.show()

```

## График 2

Постройте 4 графика на одной фигуре, как расположить координатные плоскости - это на ваше усмотрение.

Это должны быть гистограммы для следующих переменных:
- `property_type`
- `city`
- `purpose`
- `bedrooms`

Каждый график должен быть своего цвета, также задайте цвет отличный от белого для бекграунда всей фигуры.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('sales_dataset.csv')
fig, axes = plt.subplots(2, 2, figsize=(17, 11))
fig.patch.set_facecolor('#f0f0f0')

variables = ['property_type', 'city', 'purpose', 'bedrooms']
colors = ['skyblue', 'salmon', 'lightgreen', 'orange']

for ax, var, color in zip(axes.flatten(), variables, colors):
    sns.histplot(data=data, x=var, ax=ax, color=color, kde=False)
    ax.set_title(f'Гистограмма для {var}')
    ax.set_xlabel(var)
    ax.set_ylabel('Частота')

plt.tight_layout()
plt.show()
```

## График 3

Визуализируйте матрицу корреляции в виде heatmap.

Все должно быть ЧИТАБЕЛЬНО!

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

numerical_data = data.select_dtypes(include=['number'])
correlation_matrix = numerical_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Матрица корреляции')
plt.show()
```

## График 6

Постройте график среднего значения переменной `price` в зависимости от переменной `Season`.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('sales_dataset.csv')

season_mean = data.groupby('Season')['price'].mean().reset_index()
plt.figure(figsize=(8, 5))
sns.barplot(data=season_mean, x='Season', y='price', palette='viridis')
plt.title('Среднее значение price в зависимости от Season')
plt.xlabel('Season')
plt.ylabel('Средний price')
plt.show()
```

## График 7

Найдите любой **нестандартный** график и постройте его. На ваш выбор.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1.5, 1.5, 400)
y = np.linspace(-1.5, 1.5, 400)
X, Y = np.meshgrid(x, y)

F = (X**2 + Y**2 - 1)**3 - X**2 * Y**3

plt.figure(figsize=(6, 6))
plt.contour(X, Y, F, levels=[0], colors='red')
plt.axis("equal")
plt.axis("on")
plt.title("График уравнения сердца", fontsize=14)

plt.show()
```
